##### Function for pulling in players picked

```python
import pandas as pd
import requests
import json

def get_draft_players(draft_id, krusty_df):
    url = f"https://api.sleeper.app/v1/draft/{draft_id}/picks"
    response = requests.get(url)
    data = json.loads(response.text)

    # Extract first_name, last_name, and player_id, and concatenate first_name and last_name
    player_info = [(pick['metadata']['first_name'] + ' ' + pick['metadata']['last_name'],
                    pick['metadata']['player_id']) for pick in data if 'metadata' in pick]

    # Create DataFrame with the player names and ids
    players_drafted = pd.DataFrame(player_info, columns=['Player', 'player_id'])

    # Add player_pic column
    players_drafted['player_pic'] = 'https://sleepercdn.com/content/nfl/players/thumb/' + players_drafted['player_id'] + '.jpg'

    # Filter out the drafted players from krusty_df
    filtered_df = krusty_df[~krusty_df['Player'].isin(players_drafted['Player'])]

    return filtered_df
```

```python
# Merge etr and spikeweek on the ID columns
merged_df = pd.merge(spikeweek, etr, on='ID', suffixes=('_spikeweek', '_etr'))

# Since you mentioned that Position, Name, and Team columns are the same in both dataframes, we'll just keep one set
merged_df = merged_df[['ID', 'Name_spikeweek', 'Position_spikeweek', 'Team', 'rank_spikeweek', 'rank_etr', 'ADP']]
merged_df = merged_df.rename(columns={'Name_spikeweek': 'Name', 'Position_spikeweek': 'Position', 'rank_spikeweek': 'sw', 'rank_etr': 'etr'})
# First, rename 'ID' to 'Id' in the merged_df DataFrame to match the 'Id' in legup
merged_df = merged_df.rename(columns={'ID': 'Id'})

# Then, merge merged_df and legup on 'Id'
merged_df = pd.merge(merged_df, legup, on='Id', suffixes=('_merged', '_legup'))

# Continue with the remaining steps...

merged_df = merged_df.rename(columns={'Rank': 'leg'})

# Then merge with rotoviz dataframe on the Name column
final_df = pd.merge(merged_df, rotoviz, left_on='Name_merged', right_on='Name', how='left')

# Rename AVGRank to rv
final_df = final_df.rename(columns={'rank': 'rv', 'Position_x' : 'Position', 'Team_x' : 'Team'})

# If you want to remove duplicate columns (Position, Name, Team) from rotoviz, you can drop them
final_df = final_df.drop(['Position_legup', 'Team_legup'], axis=1)

# List of names
lates = ['Trey Sermon']

#Sanderson targets
jakob = ['Justin Jefferson', 'Rico Dowdle', 'Jaylen Waddle', 'Mark Andrews', 'Christian Kirk', 'Diontae Johnson', 'Brock Bowers', 'Jaylen Warren', 'Jonathan Brooks', 'Tyjae Spears', 'Brian Robinson', 'Chase Brown', 'Dontayvion Wicks', 'Xavier Legette', 'Josh Downs', 'Ja\'Lynn Polk', 'Drake Maye', 'Javon Baker', 'Demario Douglas']
gretch = ['Audric Estime', 'Juwan Johnson', 'Bo Nix', 'Jalen Tolbert', 'Ty Chandler', 'Deneric Prince', 'Bucky Irving', 'MarShawn Lloyd', 'Jalen McMillan', 'Brock Bowers', 'Josh Downs', 'Luke Musgrave', 'Roschon Johnson', 'Kyle Pitts', 'Wan\'Dale Robinson', 'J.K. Dobbins', 'Tyrone Tracy Jr.', 'Jaxon Smith-Njigba', 'Diontae Johnson']
mattek = ['Darnell Mooney','Braelon Allen', 'Jalen McMillan', 'Greg Dortch', 'Jermaine Burton', 'Joshua Palmer', 'Isaiah Likely', 'Kyler Murray']
hayden = ['Kyren Williams', 'Josh Jacobs', 'Diontae Johnson', 'James Conner', 'David Njoku', 'Pat Freiermuth', 'Rico Dowdle', 'Tyler Allgeier', 'Noah Fant', 'Johnny Wilson']
harmon = ['Romeo Doubs', 'Michael Wilson']
harmon_knocks = ['Christian Watson', 'Keenan Allen']
avoid_temp = ['Josh Downs']

#milly / spikeweek / my targets
milly = ['Andrei Iosivas', 'Darnell Mooney', 'Michael Wilson', 'Justin Fields', 'Deebo Samuel', 'Jaylen Waddle', 'Patrick Mahomes', 'Rashee Rice', 'James Cook', 'Theo Johnson', 'Jalen McMillan' 'Trey Sermon', 'Javonte Williams', 'Josh Jacobs', 'Jameson Williams', 'Curtis Samuel', 'Jayden Daniels', 'Anthony Richardson']

# Set 'rv' column to 250 for these names
#final_df.loc[final_df['Name_merged'].isin(names), 'rv'] = 250


# Define weights
weights = {
    'sw': 1.25,
    'etr': 1.5,
    'rv': 0.5,
    'leg': 2
}

final_df['rank'] = (final_df['sw'] * weights['sw'] + final_df['etr'] * weights['etr'] + final_df['rv'] * weights['rv'] + final_df['leg'] * weights['leg']) / sum(weights.values())
final_df = final_df[['Id', 'Name', 'Position', 'Team', 'ADP', 'rank']]
final_df.loc[final_df['Name'].isin(jakob), 'rank'] = final_df.loc[final_df['Name'].isin(jakob), 'rank'] * 0.90
final_df.loc[final_df['Name'].isin(gretch), 'rank'] = final_df.loc[final_df['Name'].isin(gretch), 'rank'] * 0.90
final_df.loc[final_df['Name'].isin(mattek), 'rank'] = final_df.loc[final_df['Name'].isin(mattek), 'rank'] * 0.93
final_df.loc[final_df['Name'].isin(hayden), 'rank'] = final_df.loc[final_df['Name'].isin(hayden), 'rank'] * 0.90
final_df.loc[final_df['Name'].isin(harmon), 'rank'] = final_df.loc[final_df['Name'].isin(harmon), 'rank'] * 0.90
final_df.loc[final_df['Name'].isin(harmon_knocks), 'rank'] = final_df.loc[final_df['Name'].isin(harmon_knocks), 'rank'] * 1.10
final_df.loc[final_df['Name'].isin(avoid_temp), 'rank'] = final_df.loc[final_df['Name'].isin(avoid_temp), 'rank'] * 1.50
final_df.loc[final_df['Name'].isin(milly), 'rank'] = final_df.loc[final_df['Name'].isin(milly), 'rank'] * 0.85
final_df.loc[final_df['Name'].isin(lates), 'rank'] = 225
final_df = final_df.sort_values('rank')


display(final_df)
```
